```python
import pandas as pd
import numpy as np
```

```python
ufc_ranking = pd.read_csv(r"C:\Users\Akoua Orsot\Desktop\ds_projects_data\rankings_history.csv")
ufc_ranking.head()
```

```python
champs = ufc_ranking[ufc_ranking["rank"]==0]
champs
```

```python
champs = champs.drop(columns=['date', 'rank'], axis=1)
champs
```

```python
champs.dropna(inplace=True)
champs
```

```python
champs.drop_duplicates(inplace=True)
champs
```

```python
champs.info()
```

```python
champs_R = champs.rename(columns={"fighter":"R_fighter"})
champs_B = champs.rename(columns={"fighter":"B_fighter"})
```

```python
ufc_master = pd.read_csv(r"C:\Users\Akoua Orsot\Desktop\ds_projects_data\ufc-master.csv")
ufc_master.head()
```

```python
champs_fights_R = pd.merge(champs_R, ufc_master, on="R_fighter")
champs_fights_R 
```

```python
champs_fights_B = pd.merge(champs_B, ufc_master, on="B_fighter")
champs_fights_B
```

```python
ufc_champs_final = pd.concat([champs_fights_R,champs_fights_B])
ufc_champs_final.to_csv('C:/Users/Akoua Orsot/Desktop/ds_projects_data/ufc_champs_final.csv')  
```
